# Цели и задачи

5. Найти наиболее частую категорию нарушения c аварий с виновником автомобилем, в результате которой авто не подлежит восстановлению.

Для выполнения этой задачи сначала потребуется идентифицировать случаи, в которых виновником ДТП стал автомобиль, а после получить результаты отфильтровав по признаку case_id.

```python
query = '''
SELECT case_id,
       pcf_violation_category
FROM collisions
WHERE 
    collision_damage='fatal' AND
    case_id IN 
        (SELECT case_id
        FROM parties
        WHERE at_fault=1 AND party_type='car')
'''
request_result = pd.read_sql_query(query, con = engine.raw_connection())
```

```python
sns.countplot(data=request_result, x='pcf_violation_category')
ax = plt.gca()
ax.set_title('График количества ДТП по категориям нарушений')
ax.set_ylabel('количество случаев')
ax.set_xlabel('значения признака pcf_violation_category')
ax.set_xticks(ticks=ax.get_xticks(), labels=ax.get_xticklabels(), rotation=60, ha='right')
plt.show()
```

Наибольшая вероятность получить фатальные повреждения - это водить автомобиль в нетрезвом состоянии, совершать неправильный поворот а также превышать скорость.

6. Влияет ли тип трансмиссии на вероятность аварии при парковке?

Первое - находим виновником ДТП, определяем их номер, идентификтор случая.
<br>Второе - Соединяем таблицы vehicles, полученный срез данных с таблицы parties и таблицу collisions.
<br>Фильтруем данные по признаку pcf_violation_category и соответствию виновника своему номеру.

```python
query = '''
WITH culprits as(
    SELECT
        case_id,
        party_number AS culprit_party_number
    FROM parties
    WHERE 
        party_type='car' AND
        at_fault=1)

SELECT
    v.case_id,
    v.vehicle_transmission,
    v.party_number
FROM vehicles AS v 
LEFT JOIN culprits AS cul ON v.case_id=cul.case_id
JOIN collisions AS c ON v.case_id=c.case_id
WHERE 
    v.party_number=cul.culprit_party_number AND
    c.pcf_violation_category='hazardous parking'
'''
request_result = pd.read_sql_query(query, con = engine.raw_connection())
```

```python
sns.countplot(data=request_result, x='vehicle_transmission')
ax = plt.gca()
ax.set_title('Сравнения количства аварий при парковке у автомобилей с разной трансмиссией')
ax.set_ylabel('количество случаев')
ax.set_xlabel('значения признака vehicle_transmission');
```

## Построение "скелета" модели

Поскольку одним и требований заказчика было реализовать отбор признаков на основе их статистической значимости, далее в коде будет реализована работа алгоритма `mutual_info_classif`, так как он в отличие от `chi2` способен выявлять более сложные зависимости между исследуемым признаком и целевой переменной. Поскольку все признаки кроме одного в данных категориальные, то для более корректной и точной работы алгоритма проще будет преобразовать единственный количественный признак в категориальный.

Задачу по преобразованию признака будет выполнять алгоритм кластеризации GMM (Gaussian Mixture Model). Он будет разбивать данные на заданное число кластеров. Для корректной работы в пайплайне и поддержки создадим собственный класс GMMBinner.

```python
class GMMBinner(BaseEstimator, TransformerMixin):
    def __init__(self, n_components=3, random_state=42, ohe_status=False, clip_threshold=4.0):
        self.ohe_status = ohe_status
        self.n_components = n_components
        self.random_state = random_state
        self.gmm = GaussianMixture(n_components=self.n_components, random_state=self.random_state, init_params='kmeans', n_init=1)
        self.cluster_summary = None
        self.clip_threshold = clip_threshold

    def fit(self, X, y=None):
        self.columns = X.columns
        self.gmm.fit(X.values.reshape(-1, 1))
        self.mean_ = np.mean(X.values)
        self.std_ = np.std(X.values)
        return self

    def _clip_outliers(self, X):
        lower = self.mean_ - self.clip_threshold * self.std_
        upper = self.mean_ + self.clip_threshold * self.std_
        return np.clip(X.values, lower, upper)
        
    def transform(self, X):
        try:
            cluster_prediction = self.gmm.predict(X.values.reshape(-1, 1))
        except:
            X_values = self._clip_outliers(X)
            cluster_prediction = self.gmm.predict(X_values.reshape(-1, 1))
        df = pd.concat([X.reset_index(drop=True), pd.Series(cluster_prediction, name='clusters')], axis=1)
        df = df.set_axis(['age', 'clusters'], axis=1)
        self.cluster_summary = df.groupby('clusters')['age'].agg(['min', 'max', 'mean', 'count']).sort_values('mean')
        
        if self.ohe_status:
            self.feature_names = [f"gmm_component_{i+1}" for i in range(self.n_components)]
            try:
                probs = self.gmm.predict_proba(X)
            except:
                X_values = self._clip_outliers(X)
                probs = self.gmm.predict_proba(X_values)
            preds = (probs > 0.5).astype(int)
            return pd.DataFrame(data=preds, columns=self.feature_names).iloc[:, [0, 1]]
            
        return pd.DataFrame(data=cluster_prediction.reshape(-1, 1), columns=self.columns)

    def get_feature_names_out(self, input_features=None):
        if self.ohe_status:
            return self.feature_names[:-1]
        return self.columns

    def cluster_summary(self, X):
        return self.cluster_summary

num_to_cat = Pipeline([
    ('gmm_binning', GMMBinner(n_components=3, random_state=RANDOM_STATE))
])

num_to_cat_ohe = Pipeline([
    ('gmm_binning', GMMBinner(n_components=3, ohe_status=True, random_state=RANDOM_STATE))
])

ohe_pipe = Pipeline([
    (
        'simpleimputer_ohe', 
         SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    ),
    (
        'ohe', 
        OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
    )
])

ord_pipe = Pipeline([
    (
        'simpleimputer_before_ord', 
         SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    ),
    (
        'ordinal_encoder', 
        OrdinalEncoder(handle_unknown='use_encoded_value', 
                       unknown_value=np.nan)
    ),
    (
        'simpleimputer_fater_ord', 
         SimpleImputer(missing_values=np.nan, strategy='most_frequent')
    )
])

preprocessor_tree = ColumnTransformer([
    ('num_to_cat', num_to_cat, numerical_feature),
    ('ord', ord_pipe, categorical_features)
])

preprocessor_simple = ColumnTransformer([
    ('num_to_cat', num_to_cat_ohe, numerical_feature),
    ('ohe', ohe_pipe, categorical_features)
])


class MutualInfoWithFixedRandomState:
    def __init__(self, random_state):
        self.random_state = random_state

    def __call__(self, X, y):
        return mutual_info_classif(X, y, random_state=self.random_state, n_jobs=1), None

feature_selection = GenericUnivariateSelect(
    score_func=MutualInfoWithFixedRandomState(random_state=RANDOM_STATE),
    mode='k_best',
    param=10
)



models = {
    'decision_tree': DecisionTreeClassifier(random_state=RANDOM_STATE),
    'logistic_regression': LogisticRegression(random_state=RANDOM_STATE),
    'lgbm': LGBMClassifier(verbose=-1, 
                           random_state=RANDOM_STATE, 
                           deterministic=True, 
                           feature_fraction_seed=RANDOM_STATE,
                           bagging_seed=RANDOM_STATE,
                           n_jobs=1)
}

pipelines = {}

for name, model in models.items():
    if name in ['lgbm', 'decision_tree']:
        preprocessor = preprocessor_tree
    else:
        preprocessor = preprocessor_simple
    
    pipelines[name] = Pipeline([
        ('preprocessing', preprocessor),
        ('feature_selection', feature_selection),
        ('model', model)
    ])

param_grid = {
    'decision_tree': {
        'feature_selection__param': range(5, len(categorical_features) + 2),
        'model__max_depth': range(3, 10),
        'model__min_samples_split': range(2, 10)
    },
    'logistic_regression': {
        'feature_selection__param': range(5, len(categorical_features) + 2),
        'model__C': np.linspace(0.01, 10, num=5)
    },
    'lgbm': {
        'feature_selection__param': range(5, len(categorical_features) + 2),
        'model__num_leaves': range(20, 150, 10),
        'model__max_depth': range(4, 10),
        'model__learning_rate': np.linspace(0.01, 0.1, num=10),
        'model__n_estimators': range(300, 1000, 100)
    }
}
```

В качестве метрики оценки модели будет выбрана ROC-AUC, так как она оценивает модель на всех возможных порогах классификации, а не на одном фиксированном значении. Это важно, потому что в зависимости от задачи и ситуации, порог принятия решения может изменяться.

```python
searches = {}

for name, pipeline in pipelines.items():
    print(f"Запускаем RandomizedSearchCV для модели: {name}")
    set_seed()
    search = RandomizedSearchCV(
        clone(pipeline),
        param_distributions=param_grid[name],
        n_iter=10,
        cv=3,
        scoring='roc_auc',
        random_state=RANDOM_STATE
    )
    search.fit(X_train, y_train)
    searches[name] = search
```

Опишем полученные результаты, адаптировав их к текущей задаче.

TP (True Positive) – Водители, которых модель правильно определила как виновных в ДТП.
<br>FN (False Negative) – Водители, которых модель не смогла выявить как виновных (пропустила опасных клиентов).
<br>FP (False Positive) – Водители, которых модель ошибочно определила как виновных в ДТП.
<br>TN (True Negative) – Водители, которых модель правильно определи как безопасных.

* *Метрика **precision** (точность)* - показывает сколько клиентов были действительно виновными из тех, которых предсказала модель.
* *Метрика **recall** (полнота)* - показывает сколько из всех реально виновных водителей сколько система смогла выявить.

Выбор интересующей метрики может меняться в зависимости от интересов компании и бизнес задач, но для минимизации ДТП нам необходимо оптимизировать recall (полноту), поскольку её увеличение означает, что система минимизирует пропуск опасных водителей, снижая будущие убытки.
